 n이 홀수라면 리프 노드 개수는 (n+1)/2, 내부 노드 개수는 (n-1)/2이다. 즉, 리프 노드 개수 = 내부 노드 개수+1

귀납적 가정: k < h인 모든 k에 대해 성립한다 가정하자.
<증명>
귀납적 전개: (정리가 k = h에 대해 성립함을 보인다)
이진 검색 트리 T는 루트 노드를 r로 하며, 왼쪽 서브 트리에 TL, 오른쪽 서브 트리에 TR을 둘 수 있다.
귀납적 가정에 의해 중위 순회는 TL과 TR을 각각 정렬된 순서로 방문한다.
TL의 모든 키는 r의 키보다 작고, TR의 모든 키는 r의 키보다 크므로,
중위 순회 TL→ r → TR는 정렬된 순서다.
